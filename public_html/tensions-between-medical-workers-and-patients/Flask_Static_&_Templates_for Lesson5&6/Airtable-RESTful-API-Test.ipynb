```python
#
# Retrieve records in a list
#
import requests
import pandas as pd

headers = {
        'Authorization': 'Bearer your-API-key',
}

r = requests.get('https://api.airtable.com/v0/appM38HXlEVhxmnqx/flaskdemo?sortField=_createdTime&sortDirection=desc', headers=headers)
print("Status Code:",r.status_code)
print(r.text)
dict = r.json()
a_list = dict['records']
df = pd.DataFrame.from_dict(a_list)
df.head()
```

```python
#
# Create record
#

import requests
import pandas as pd

fname = "Tony"
lname = "Choi"
student_id = "786868"
date_of_birth = "2002-02-21"

fields =  {
    "fname": fname,
    "lname": lname,
    "student_id": student_id,
    "date_of_birth": date_of_birth
    }

data = {"fields": fields}

headers = {'Authorization': 'Bearer Bearer your-API-key', 'Content-Type': 'application/json; charset=utf-8'}
r = requests.post('https://api.airtable.com/v0/appM38HXlEVhxmnqx/flaskdemo',json=data,headers=headers)
print("Status Code:",r.status_code)
print(r.text)
dict = r.json()
df = pd.DataFrame.from_dict(dict)
df.head()
```
